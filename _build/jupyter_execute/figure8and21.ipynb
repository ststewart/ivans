## Ideal dust-gas mixture equation of state

The model equation of state is taken from {cite}`Steiner_Hirschler_2002`.

The silicate specific heat capacity is for olivine at high temperature from {cite}`su2018`.


```python
import numpy as np
import matplotlib.pyplot as plt

from matplotlib.collections import LineCollection
from matplotlib.colors import ListedColormap, BoundaryNorm

import holoviews as hv
from holoviews import opts
hv.extension('bokeh')

from bokeh.plotting import figure, show
from bokeh.io import output_notebook

import colormaps as local_cmaps

#https://gist.github.com/thriveth/8560036
CBcol = ['#377eb8', '#ff7f00', '#4daf4a',
                  '#f781bf', '#a65628', '#984ea3',
                  '#999999', '#e41a1c', '#dede00']
#https://github.com/mpetroff/accessible-color-cycles
CBcol = ["#1845fb", "#ff5e02", "#c91f16", "#c849a9", "#adad7d", "#86c8dd", "#578dff", "#656364"]
output_notebook()
```

```python
rhogas = 2.e-7 # kg/m3
rhodust = 3000 # kg/m3
vol = 1 # m3
mgas = rhogas*vol # kg
dtog = np.asarray([.01,0.1,1,10,100,1000,10000,100000]) # = mdust/mgas
#dtog = np.asarray([10]) # = mdust/mgas
#dtog = np.asarray([10000]) # = mdust/mgas
#dtog = np.asarray([1]) # = mdust/mgas
mdust = mgas*dtog # kg
m = mgas + mdust
rho = m/vol
k = mdust/m # = mdust/mtot
vdust = mdust/rhodust
# v = vgas + vdust
G = rhodust/rhogas
# vanishing dust volume G -> infinity
Z = vdust/vol 
Z2 = k / ( (1-k)*G+k )
#print(Z,Z2,G)
#print(rho)
#print(Z/Z2)
#print(dtog,Z,k)
```

```python
T=200 # K
Rgas=8.314 # J/mol/K
#moltokg = .002 # pure H2 gas kg/mol
fh=12 # lodders elemental abundances for proto sun
fhe=10.984
mugas = (fh*1+fhe*4)/(fh+fhe)
#print(mugas)
moltokg = .002433 # H2+He gas kg/mol
```

```python
# PV = nRT # J
# J/mol/K (mol/kg) (kg/m3) K   =  J/m3
#P = (1-k)/(1-Z) * Rgas/moltokg*rhogas*T # a partial pressure
P = (1-k)/(1-Z) * Rgas/moltokg*rho*T # a partial pressure
Pidg = Rgas/moltokg*rhogas*T
#print(P,T,rhogas,Pidg)
#print(P/Pidg)
```

```python
delta = k/(1-k)
beta = 1000/(5/2*Rgas/moltokg)
g = 1.4
gamma = (g+delta*beta)/(1+delta*beta)
cs = np.sqrt( gamma*(1-k)/(1-Z)/(1-Z) * Rgas/moltokg * T)
ie = (1-Z)/(gamma-1)*(P/rho)
cvgas = 5/2*Rgas/moltokg
cpgas = 1.4*cvgas
cpdust = 1000 # e.g. olivine at a few 100 K http://english.gyig.cas.cn/pu/papers_CJG/201803/P020180316540125048801.pdf
cpmix = k*cpdust+(1-k)*cpgas
cvmix = cpmix/gamma
#print(delta,beta,gamma, cs, ie)
#print(cvgas,cvmix,ie,gamma,rho)
```

```python

figure, ax = plt.subplots(3, 2,figsize=(8,10)) 
i,j=0,0
ax[i,j].plot(dtog,cs,'-x')
#ax[i,j].plot(dtog,np.sqrt(gamma*P/rho),'k-d') # ideal gas sound speed
ax[i,j].set_xlabel('Dust to gas mass ratio')
ax[i,j].set_ylabel('Sound speed [m/s]')
ax[i,j].set_xscale('log')
ax[i,j].set_yscale('log')
#ax[i,j].set_title('Dusty-ideal-gas model T=200 K')

i,j=0,1
ax[i,j].plot(dtog,P,'-x')
ax[i,j].set_xlabel('Dust to gas mass ratio')
ax[i,j].set_ylabel('Pressure [Pa]')
ax[i,j].set_xscale('log')
ax[i,j].set_yscale('log')

i,j=1,0
ax[i,j].plot(dtog,ie,'-x')
ax[i,j].set_xlabel('Dust to gas mass ratio')
ax[i,j].set_ylabel('Internal energy [J/kg]')
ax[i,j].set_xscale('log')
ax[i,j].set_yscale('log')

i,j=1,1
ax[i,j].plot(dtog,rho,'-x')
ax[i,j].set_xlabel('Dust to gas mass ratio')
ax[i,j].set_ylabel('Bulk density [kg/m$^3$]')
ax[i,j].set_xscale('log')
ax[i,j].set_yscale('log')

i,j=2,0
ax[i,j].plot(dtog,vdust/vol,'-x')
ax[i,j].set_xlabel('Dust to gas mass ratio')
ax[i,j].set_ylabel('Dust volume fraction')
ax[i,j].set_xscale('log')
ax[i,j].set_yscale('log')

i,j=2,1
ax[i,j].plot(dtog,cvmix,'-x')
ax[i,j].set_xlabel('Dust to gas mass ratio')
ax[i,j].set_ylabel('Bulk sp. heat capacity $c_v$ [J/K/kg]')
ax[i,j].set_xscale('log')
ax[i,j].set_yscale('log')

figure.tight_layout()
#plt.savefig('dusty-gas.pdf')
plt.show()
```

```python

nr=len(rho)
npres=1000
hugallp = np.zeros((nr,npres))
hugallr = np.zeros((nr,npres))
hugallt = np.zeros((nr,npres))
hugallup = np.zeros((nr,npres))
hugallus = np.zeros((nr,npres))

#for ir in range(5,len(rho)):
for ir in range(0,7):
    rarr = np.power(10,np.arange(0,5,.01))*rho[ir]
    parr = np.power(10,np.arange(0,10,.01))*P[ir]
    #print(len(parr))
#    print(rarr)
#    print(parr)
    e0 = (1-Z[ir])/(gamma[ir]-1)*(P[ir]/rho[ir])
    rhug = np.zeros(0)
    #print(rho[ir],P[ir],e0)
    for ip in range(len(parr)):
        earr = (1-Z[ir])/(gamma[ir]-1)*(parr[ip]/rarr)
        huge = (earr-e0)-0.5*parr[ip]*(1/rho[ir]-1/rarr)
        rint = np.interp(0,np.flip(huge),np.flip(rarr))
        eint = np.interp(rint,rarr,earr)
        hugallp[ir,ip]=parr[ip]
        hugallr[ir,ip]=rint
        hugallt[ir,ip]=eint/cvmix[ir]
        hugallup[ir,ip] = np.sqrt((hugallp[ir,ip]-P[ir])*(1/rho[ir]-1/hugallr[ir,ip]))
        hugallus[ir,ip] = (1/rho[ir])*np.sqrt((hugallp[ir,ip]-P[ir])/(1/rho[ir]-1/hugallr[ir,ip]))
        #print(huge,rarr,rint)
        #stop
```

```python

figure, ax = plt.subplots(3, 2, figsize=(13,13)) 

plt.rcParams.update({'font.size': 16})

i,j=0,0
ax[i,j].plot(dtog,cs,'-x')
#ax[i,j].plot(dtog,np.sqrt(gamma*P/rho),'k-d') # ideal gas sound speed
ax[i,j].set_xlabel('Dust to gas mass ratio')
ax[i,j].set_ylabel('Sound speed (m/s)')
ax[i,j].set_xscale('log')
ax[i,j].set_yscale('log')
#ax[i,j].set_title('Dusty-ideal-gas model T=200 K')
ax[i,j].text(1.e5,800,'A')

i,j=0,1
ax[i,j].plot(dtog,cvmix,'-x')
ax[i,j].set_xlabel('Dust to gas mass ratio')
ax[i,j].set_ylabel('Sp. heat capacity $c_v$ (J/K/kg)')
ax[i,j].set_xscale('log')
ax[i,j].set_yscale('linear')
ax[i,j].text(1.e5,8000,'B')

i,j=1,0
ax[i,j].plot(dtog,ie/1.e6,'-x')
ax[i,j].set_xlabel('Dust to gas mass ratio')
ax[i,j].set_ylabel('Sp. Internal energy (MJ/kg)')
ax[i,j].set_xscale('log')
ax[i,j].set_yscale('linear')
ax[i,j].text(1.e5,1.6,'C')

i,j=1,1
ax[i,j].plot(dtog,vdust/vol,'-x')
ax[i,j].set_xlabel('Dust to gas mass ratio')
ax[i,j].set_ylabel('Dust volume fraction')
ax[i,j].set_xscale('log')
ax[i,j].set_yscale('log')
ax[i,j].text(1.e5,2.e-6,'D')

i,j=2,0
ax[i,j].plot(dtog,rho,'-x')
ax[i,j].set_xlabel('Dust to gas mass ratio')
ax[i,j].set_ylabel('Mixture density (kg/m$^{3}$)')
ax[i,j].set_xscale('log')
ax[i,j].set_yscale('log')
ax[i,j].text(1.e5,.8e-2,'E')

i,j=2,1
ax[i,j].plot(hugallr[0,:],hugallp[0,:],label='$m_d$/$m_g$='+str(dtog[0]))
ax[i,j].plot(hugallr[1,:],hugallp[1,:],label=dtog[1])
ax[i,j].plot(hugallr[2,:],hugallp[2,:],label=dtog[2])
ax[i,j].plot(hugallr[3,:],hugallp[3,:],label=dtog[3])
ax[i,j].plot(hugallr[4,:],hugallp[4,:],label=dtog[4])
ax[i,j].plot(hugallr[5,:],hugallp[5,:],label=dtog[5])
ax[i,j].plot(hugallr[6,:],hugallp[6,:],label=dtog[6])
ax[i,j].set_xlabel('Mixture Density (kg m$^{-3}$)')
ax[i,j].set_ylabel('Pressure (Pa)')
ax[i,j].set_xscale('log')
ax[i,j].set_yscale('log')
ax[i,j].set_ylim(.1,1.e5)
ax[i,j].text(12,2.e4,'F')

ax[i,j].legend(fontsize=10)

#ax[2, 1].axis("off")

figure.tight_layout()
plt.savefig('./plots/Fig21-dusty-gas-model-paper.pdf')
plt.show()
```
